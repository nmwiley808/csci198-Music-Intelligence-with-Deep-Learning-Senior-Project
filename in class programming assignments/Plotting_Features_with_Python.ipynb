```python
query = """
select sepal_length,	sepal_width,	petal_length,	petal_width,	species from iris;
"""
iris_df = pd.read_sql_query(query, NEON_CON_IRIS)

iris_df.head()
```

# Bar Chart Interpretation:

This bar chart shows the number of observations for each Iris species in the dataset. All three species, setosa, versicolor, and virginica, appear in equal amounts, indicating that datset is balanced. A balanced dataset is important because it prevents bias toward one category when analyzing patterns or building predictive models.

---

```python
# Bar Chart - Count of Each Species
species_counts = iris_df['species'].value_counts()

plt.figure()
species_counts.plot(kind='bar')

plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')

plt.show()
```

# Histogram Interpretation:

This histogram displays the distribution of sepal length values across all Iris samples. The data appears to be roughly normally distributed, with most sepal lengths clustered around the middle range. This suggests that extreme values are less common and that sepal length varies within a relatively consistent range.

---

```python
# Histogram - Distribution of Sepal Length
plt.figure()
plt.hist(iris_df['sepal_length'], bins=10)

plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.show()
```

# Scatter Plot Interpretation:

This scatter plot illustrates the relationship between sepal length and petal length for each Iris species. The species form noticeable clusters, especially setosa, which is clearly separated from the others. This indicates that petal and sepal measurements are useful features for distinguishing between species.

---

```python
# Scatter Plot - Sepal Length vs Petal Length by Species
species_list = iris_df['species'].unique()

plt.figure()
for s in species_list:
    subset = iris_df[iris_df['species']== s]
    plt.scatter(subset['sepal_length'],
                subset['petal_length'],
                       label=s)

plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()

plt.show()
```
