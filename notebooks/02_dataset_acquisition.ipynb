## 02 – Dataset Acquisition (Hugging Face GTZAN)

This notebook loads the GTZAN dataset from Hugging Face

```python
from datasets import load_dataset

gtzan = load_dataset("marsyas/gtzan")

print(gtzan)
```

```python
sample = gtzan["train"][0]

print("Keys:", sample.keys())
print("Genre:", sample["genre"])

audio = sample["audio"]
print("Sampling rate:", audio["sampling_rate"])
print("Audio shape:", audio["array"].shape)
```

```python
print("Total samples:", len(gtzan["train"]))
```

```python
import pandas as pd

metadata = []

for i in range(len(gtzan["train"])):
    sample = gtzan["train"][i]
    metadata.append({
        "index": i,
        "genre": sample["genre"],
        "sampling_rate": sample["audio"]["sampling_rate"],
        "num_samples": len(sample["audio"]["array"])
    })

df = pd.DataFrame(metadata)

df.head()
```

```python
METADATA_PATH = "/content/drive/MyDrive/csci198-data/metadata/gtzan_metadata.csv"

df.to_csv(METADATA_PATH, index=False)

print("Metadata saved to:", METADATA_PATH)
```
