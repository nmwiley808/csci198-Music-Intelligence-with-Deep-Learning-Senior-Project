```python
# Test Audio Provessing
import librosa
import numpy as np

sr = 22050
t = np.linspace(0, 1, sr)
y = 0.5 * np.sin(2 * np.pi * 440 * t)

mel = librosa.feature.melspectrogram(y=y, sr=sr)
print("Mel Spectrogram Shape:", mel.shape)
```
